```
from bs4 import BeautifulSoup
import datetime
import requests


def process_page(url):
  
  response = requests.get(url)
  
  if(response.ok):
    
    content = BeautifulSoup(response.text, 'html.parser').select('.content')
    
    for pieces in content:
      
      #Scrape for article title
      title = pieces.select_one('header > h3').get_text()
      print(title)
      
      #Scrape for article author
      author = pieces.select_one('header > p').get_text()
      print(author[4:].strip())  #Get rid of "By"
      
      #Scrape for article date
      Date = pieces.select_one('header > .byline > time')['datetime']
      print(str(Date)[0:10])   #Just show the date,  not the time
      
      #Print article synopsis
      synopsis = pieces.select_one('p.synopsis').get_text()
      print(synopsis.strip())
      
      # Separator for readability
      print('----------------------------------------------------------------------------------')
      
  else:
    
    print('URL '+url+' might be incorrect. Have you checked that everything is spelled correctly?')

    
    
process_page('http://www.space.com/news')
```
