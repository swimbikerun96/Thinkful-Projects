```
#2. Are there any years with an unusually high number of releases? How do you define unusually high? Unusal is equal to 2 standard deviations greater than the mean.
movies = engine.execute('WITH cte as (SELECT COUNT(movie_id) AS quantity, EXTRACT(YEAR FROM movies.release_date) AS release_year FROM movies GROUP BY release_year) SELECT release_year, quantity FROM cte WHERE quantity NOT BETWEEN (SELECT AVG(quantity) - (2*STDDEV(quantity)) from cte) AND (SELECT AVG(quantity) + 2*STDDEV(quantity) from cte);')

engine.dispose()
for row in movies:
  print(row)
```

```
#4. What is the median movie length?

import math

sql = '''
select 
  runtime
from
  movies
where runtime IS NOT null'''

runtime = engine.execute(sql)
engine.dispose()
rows = runtime.fetchall()

runtime = [x['runtime'] for x in rows]
print(runtime)

def median(numbers):
  numbers.sort()
  count = len(numbers)
  isEven = count %2 == 0
  
  if (isEven):
    mid = math.floor(count/2)
    a  = numbers[mid]
    b = numbers[mid + 1]
    ans = (a + b) / 2
  else:
    ans = numbers[math.floor(count/2)]
    
  return ans
    
print('The median of the runtimes is {}'.format(median(runtime)))
```
